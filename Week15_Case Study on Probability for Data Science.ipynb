```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("mushrooms.csv")
data.head()
```

```python
data.info()
```

```python
rows,cols = data.shape
mush = data[data['class']=='e'].count()["class"]
avg = mush/rows*100
print(f'Total number of Mushrooms that are Edible: {mush}\n')
print(f'Percentage of Edible Mushrooms in the dataset are {avg:.3g}%\n')

mush = data[data['class']=='p'].count()["class"]
avg = mush/rows*100

print(f'Total number of Mushrooms that are Poisonous: {mush}')
print(f'Percentage of Poisonous Mushrooms in the dataset are {avg:.3g}%')
```

```python
data.describe()
```

```python
X=data.drop('class',axis=1) 
y=data['class'] 
X.head()
```

```python
#LabelEncoding

from sklearn.preprocessing import LabelEncoder
Encoder_X = LabelEncoder() 
for col in X.columns:
    X[col] = Encoder_X.fit_transform(X[col])
Encoder_y=LabelEncoder()
y = Encoder_y.fit_transform(y)
X.head()
```

```python
X=pd.get_dummies(X,columns=X.columns,drop_first=True)
X.head()
```

```python
#Splitting the data set into training and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#feature Scaling

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train.shape
```

```python
X_test.shape
```

Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,f1_score
logit_model=LogisticRegression(random_state=42,max_iter=1000)
logit_model.fit(X_train,y_train)
y_pred=logit_model.predict(X_test)
print("Accuracy is:",round(accuracy_score(y_test,y_pred),3))
print("f1 score is:",round(f1_score(y_test,y_pred),3))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train,y_train)
print("Test Accuracy: {}%".format(round(dt.score(X_test,y_test)*100,2)))
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(X_train,y_train)
print("Test Accuracy: {}%".format(round(nb.score(X_test,y_test)*100,2)))
```

Conclusion: classfication methods give 100% accuracy with this dataset.
